# Backpropagation

## Mathematical formula

### Sigmoid Function:
The sigmoid function, often used as an activation function in neural networks, is defined as:

$\sigma(x) = \frac{1}{1 + e^{-x}}$


The sigmoid function maps any input value to a value between 0 and 1, making it useful for binary classification problems.

### Derivative of the Sigmoid Function:
The derivative of the sigmoid function is:
$\sigma'(x) = \frac{e^{-x}}{(1 + e^{-x})^2}$

This derivative can also be expressed in terms of the sigmoid function itself:
$\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))$

The sigmoid function is an S-shaped curve that maps any input value to a value between 0 and 1. Its derivative, on the other hand, has its maximum value at x = 0 and approaches zero as x goes to positive or negative infinity.

### Backpropagation:
Backpropagation is an algorithm used to compute the gradient of the loss function with respect to each weight in the neural network. The core of backpropagation is the application of the chain rule from calculus. For a given weight w  in the network:

$\frac{\partial L}{\partial w} = \frac{\partial L}{\partial a} \times \frac{\partial a}{\partial z} \times \frac{\partial z}{\partial w}$

Where:
- L is the loss function.
- a is the activation after applying the sigmoid function.
- z is the weighted sum of inputs before activation.
- $\frac{\partial L}{\partial a}$ is the gradient of the loss with respect to the activation.
- $\frac{\partial a}{\partial z}$ is the derivative of the sigmoid function, which is $\sigma(x) \times (1 - \sigma(x))$.
- $\frac{\partial z}{\partial w}$ is the input to the neuron.

The backpropagation algorithm computes these gradients for every weight in the network, starting from the output layer and working backward to the input layer.

## Python Implementation
### Sigmoid function and its derivative
Now, let's implement and plot the sigmoid function and its derivative in Python using `matplotlib` and `numpy`.
Here's the visualization of the sigmoid function and its derivative:

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

x = np.linspace(-10, 10, 400)
y = sigmoid(x)
y_prime = sigmoid_derivative(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sigmoid Function', color='blue')
plt.plot(x, y_prime, label='Derivative of Sigmoid', color='red', linestyle='--')
plt.title('Sigmoid Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

- The blue curve represents the sigmoid function, $\sigma(x) = \frac{1}{1 + e^{-x}}$. As you can see, it maps any input value to a value between 0 and 1.
  
- The red dashed curve represents the derivative of the sigmoid function, $\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))$. The derivative has its maximum value at x = 0 and approaches zero as x goes to positive or negative infinity.

### Backpropagation

**Concept**: Backpropagation is an optimization algorithm used for minimizing the error in a neural network. It works by computing the gradient of the loss function with respect to each weight by the chain rule and then updating the weights of the network in the opposite direction of the gradient.

#### Steps:
1. **Forward Pass**: Compute the output of the neural network given an input.
2. **Compute the Loss**: Calculate the difference between the predicted output and the actual label.
3. **Backward Pass**: Compute the gradient of the loss with respect to each weight.
4. **Update Weights**: Adjust the weights in the direction that reduces the error.

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Sample data
inputs = np.array([0.5, 0.6, 0.1])
weights = np.array([0.1, 0.2, 0.3])
target = 1  # True label

# Learning rate
eta = 0.01

# Forward pass
z = np.dot(inputs, weights)
a = sigmoid(z)

# Compute loss (mean squared error)
# reason for 0.5 is simplified version for single data points
loss = 0.5 * (a - target) ** 2

# Backward pass
delta = (a - target) * sigmoid_derivative(z)
gradient = delta * inputs

# Update weights
weights = weights - eta * gradient

print(f"Loss: {loss}")
print(f"Gradient: {gradient}")
print(f"Updated weights: {weights}")
```

- Add the iteration 10 times

```python
import numpy as np

# Define the sigmoid function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Sample data
inputs = np.array([0.5, 0.6, 0.1])
weights = np.array([0.1, 0.2, 0.3])
target = 1  # True label
learning_rate = 0.01

# Iterate 10 times
for iteration in range(10):
    # Forward pass
    z = np.dot(inputs, weights)
    a = sigmoid(z)

    # Compute loss (mean squared error)
    loss = 0.5 * (a - target) ** 2

    # Backward pass
    delta = (a - target) * sigmoid_derivative(z)
    gradient = delta * inputs

    # Weight update
    weights -= learning_rate * gradient

    # Print the loss and updated weights
    print(f"Iteration {iteration+1}:")
    print(f"Loss: {loss}")
    print(f"Gradient: {gradient}")
    print(f"Updated Weights: {weights}\n")
```

For simplicity, let's consider a single-layer neural network (perceptron) with a sigmoid activation function. We'll implement the forward and backward passes and visualize the gradient computation.


In the above code:
- `sigmoid` is the activation function.
- `sigmoid_derivative` computes the derivative of the sigmoid function.
- The forward pass computes the output of the perceptron.
- The backward pass computes the gradient of the loss with respect to the weights using the chain rule.

In the context of neural networks, the backward pass refers to the process of computing the gradient of the loss function with respect to each parameter (in this case, the weights) and updating the parameters in the direction that reduces the loss. This process is often referred to as backpropagation.   

Let's break down the steps of the backward pass in the code you provided, along with the corresponding mathematical equations.   

1. **Forward Pass**: First, the neural network makes a prediction (`a`) using the current weights and the sigmoid activation function.   

   - The weighted sum `z` is computed by the dot product of `inputs` and `weights`:   
     $z = \mathbf{inputs} \cdot \mathbf{weights}$   
     In this case, `z` is a scalar because it's the sum of products of the corresponding elements.   

   - The prediction `a` is the output of the sigmoid function applied to `z`:   
     $a = \sigma(z)$   
     $\sigma(z) = \frac{1}{1 + e^{-z}}$   

2. **Loss Calculation**: The loss is calculated using the mean squared error, which for a single data point is just the squared difference between the prediction and the target, scaled by 0.5 to make the derivative simpler:
   $\text{loss} = \frac{1}{2}(a - \text{target})^2$

3. **Backward Pass**: The goal is to compute the gradient of the loss with respect to the weights, which tells us how to adjust the weights to decrease the loss.   

   - The first step is to compute the derivative of the loss with respect to the prediction `a`. This is given by the chain rule:   
     $\frac{\partial \text{loss}}{\partial a} = (a - \text{target})$

   - Then, we need to compute the derivative of the prediction `a` with respect to `z`, which is the derivative of the sigmoid function:   
     $\frac{\partial a}{\partial z} = \sigma(z) \cdot (1 - \sigma(z))$   
     This comes from the derivative of the sigmoid function.   

   - The chain rule gives us the derivative of the loss with respect to `z` (also known as `delta` in backpropagation):   
     $\delta = \frac{\partial \text{loss}}{\partial z} = \frac{\partial \text{loss}}{\partial a} \cdot \frac{\partial a}{\partial z} = (a - \text{target}) \cdot \sigma(z) \cdot (1 - \sigma(z))$   

   - Finally, to update the weights, we need to compute the gradient of the loss with respect to each weight. This is again done using the chain rule, which gives us the gradient for each weight as the product of `delta` and the corresponding `input`:      
     $\frac{\partial \text{loss}}{\partial \mathbf{weights}} = \delta \cdot \mathbf{inputs}$   
     This is the gradient that is used to update the weights in gradient descent.   

4. **Weight Update**: The actual update of the weights is not shown in the code snippet you provided, but it would typically be performed as follows:  
   $\mathbf{weights} = \mathbf{weights} - \eta \cdot \frac{\partial \text{loss}}{\partial \mathbf{weights}}$   
   where $\eta$ is the learning rate, a small positive scalar determining the step size of the update.

In summary, the backward pass is a step-by-step process where we apply the chain rule multiple times to find out how the loss changes with respect to each weight, and then we use this information to update the weights in a direction that minimizes the loss.


This is a simple example, but it captures the essence of backpropagation. In practice, frameworks like TensorFlow and PyTorch handle these computations efficiently for deep neural networks.

## Examples in Machine Learning

### Backpropagation
**Python Example**: Training a simple neural network using TensorFlow/Keras for binary classification.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Generate some dummy data
np.random.seed(0)
data = np.random.randn(1000, 20)
labels = (np.sum(data, axis=1) > 0).astype(int)

# Split into training and test sets
train_data = data[:800]
train_labels = labels[:800]
test_data = data[800:]
test_labels = labels[800:]

# Build a simple neural network model
model = keras.Sequential([
    keras.layers.Dense(10, activation='relu', input_shape=(20,)),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model using backpropagation
model.fit(train_data, train_labels, epochs=10, batch_size=32)

# Evaluate the model
test_loss, test_acc = model.evaluate(test_data, test_labels)
print(f"Test Accuracy: {test_acc * 100:.2f}%")
```


In the above example, the `model.fit` function trains the neural network using backpropagation. The optimizer `adam` adjusts the weights based on the computed gradients.

### Usage in Machine Learning:

1. **Deep Learning Models**: Almost all deep learning models, including CNNs (Convolutional Neural Networks), RNNs (Recurrent Neural Networks), and transformers, use backpropagation for training.
  
2. **Transfer Learning**: When fine-tuning pre-trained models on a specific task, backpropagation is used to update the weights of the model for the new task.

3. **Reinforcement Learning**: Neural networks used in reinforcement learning, such as those in Deep Q-Learning, also utilize backpropagation.

4. **Generative Adversarial Networks (GANs)**: Both the generator and discriminator in GANs are trained using backpropagation.

Backpropagation is the backbone of training deep learning models. Understanding its mechanics helps in debugging, creating custom models, and optimizing neural networks for various tasks.
